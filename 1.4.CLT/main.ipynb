```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import expon, norm
from math import exp, sqrt, pi
```

# Строим теоретический график распредления рэлея и гистрограмму по данному распределению

```python
scale = 1
s = np.random.rayleigh(scale, 1000)
```

```python
count, bins, ignored = plt.hist(s, 30, density=1, label='Sample')
x = np.arange(0, 4, 0.1)
pdf = x / scale**2 * np.exp((-x * x) / (2 * scale**2))
plt.plot(x, pdf, label='Rayleigh theory')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

```python
print(s.mean())
```

# Генерация выборки из пяти значений

```python
mean = []
n = 5
for i in range(1000):
    s = np.random.rayleigh(scale, n)
    mean.append(s.mean())
```

## Подсчет значения нормального распределения

```python
mu = scale * sqrt(2 * pi) / 2
dispersion = (4 - pi) / 2 * scale ** 2
sigma = sqrt(8 - 2 * pi) / 2 * scale

clt = (sigma) / (scale * sqrt(n))
print(mu, dispersion, sigma, clt)
```

```python
count, bins, ignored = plt.hist(mean, n, density=1, label='Sample')
x = np.arange(0, 4, 0.1)
pdf = x / scale**2 * np.exp((-x * x) / (2 * scale**2))
plt.plot(x, pdf,label='Rayleigh theory')

norm_rv = norm(mu, clt)
plt.plot(x, norm_rv.pdf(x), label='Norm theory')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()

plt.show()
```
